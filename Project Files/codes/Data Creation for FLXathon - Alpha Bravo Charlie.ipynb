### Steps to create Flexi Rate Data

The Underwriting data created in the first step was used to train a GLM/Forest Model. Basis the champions model output, the flexible rate data was created. 

You will notice in the below data being imported there are columns "P_Achieved_Premium', "EM_PREDICTION", which are model outputs

```python
flexi_rate = pd.read_csv(r'C:\Users\sinnsr\OneDrive - SAS\FLXathon 2020\datasets\flexi_data_latest.csv') #insert your data path here
```

```python
flexi_rate.columns
```

We created 3 discount bands for creation of our training data. The discounts were logically chosen basis the Occupation, Car Type, Car Age etc. You can further modify this to suit your usecase.

```python
top_disc = [-45,-46,-47,-48,-49,-50]
med_disc = [-38,-39,-40,-41,-42,-43,-44]
low_disc = [-30,-31,-32,-33,-34,-35,-36,-37]
final_value = []
import random
```

```python
flexi_rate['Occupation'].unique()
```

```python
flexi_rate['Car_Type'].unique()
```

```python
for i in range(len(flexi_rate)):
    
    car_type = flexi_rate['Car_Type'][i]
    car_age = flexi_rate['Car_Age'][i]
    occupation = flexi_rate['Occupation'][i]
    #print(car_type,terrain_type,no_claims,car_age)
    v = 0
    #print('here')
    
    if car_type in ['SUV','Sports Car']:
        if occupation in ['Blue Collar','Lawyer','Doctor']:
        #print('first')
            if car_age <=2:
                v = random.choice(top_disc)/100
                final_value.append(v)
            else:
                v = random.choice(med_disc)/100
                final_value.append(v)
        else:
            v = random.choice(low_disc)/100
            final_value.append(v)            
            
    elif car_type in ['Minivan', 'Van', 'Panel Truck','Pickup']:
        if occupation in ['Blue Collar','Lawyer','Doctor']:
            if car_age <=2:
                v = random.choice(top_disc)/100
                final_value.append(v)
            else:
                v = random.choice(med_disc)/100
                final_value.append(v)   
        else:
            v = random.choice(low_disc)/100
            final_value.append(v)
    else:
        final_value.append(0)
```

```python
flexi_rate['Discounts'] = final_value
```

```python
flexi_rate['Flexi Rate'] = flexi_rate['P_Achieved_Premium'] + flexi_rate['P_Achieved_Premium']*flexi_rate['Discounts']
```

```python
flexi_rate.to_csv('flexi_data.csv',index=False)
```

#### Whether to load or discount is driven by Car_Type, Terrain_Type, No of claims, Car Age

```python
load_list = [3,4,5,6,7]
disc_list = [-3,-4,-5,-6,-7]
final_value = []
```

```python
for i in range(len(filtered_df)):
    
    car_type = filtered_df['Car_Type'][i]
    terrain_type = filtered_df['Terrains'][i]
    no_claims   = filtered_df['Clm_Freq'][i]
    car_age = filtered_df['Car_Age'][i]
    tsp = filtered_df['terrain_statistical_premium'][i]
    
    #print(car_type,terrain_type,no_claims,car_age)
    
    v = 0
    #print('here')
    
    if car_type in ['SUV','Pickup','Panel Truck'] and terrain_type in ['Mountain','Forest','Highway']:
        #print('first')
        if no_claims > 3 and car_age >=5:
            v = tsp + tsp*random.choice(load_list)/100
            final_value.append(v)
        else:
            v = tsp + tsp*random.choice(disc_list)/100
            final_value.append(v)                
            
    elif car_type in ['Sports Car','Minivan'] and terrain_type in ['Mountain','Forest']:
        v = tsp + tsp*0.1
        final_value.append(v)
        
    elif car_type in ['Sports Car','Minivan'] and terrain_type in ['Highway']:
        v = tsp - tsp*0.1
        final_value.append(v)
        
    elif car_type in ['Van'] and terrain_type in ['Mountain','Forest','Highway']:
        #print('inside Van')
        v = tsp + tsp*random.choice(disc_list)/100
        final_value.append(v)
    else:
        final_value.append(0)
```

```python
filtered_df['add_on_premium'] = final_value
```
